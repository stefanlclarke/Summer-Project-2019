```python
import numpy as np
import torch
import torch.nn as nn
from training import train
from NP import NP
import matplotlib.pyplot as plt
```

```python
def f(point, alpha):
    return np.sin(point*alpha)

num_datapoints = 1000
alpha_range = [-2, 2]
x_dimension = 1
y_dimension = 1

mesh = torch.linspace(-2, 2, 100).unsqueeze(0).transpose(0,1)
datapoints = []
for _ in range(num_datapoints):
    num = np.random.uniform(alpha_range[0], alpha_range[1])
    value = f(mesh, num)
    datapoints.append([mesh, value])

points = np.array([17,  5, 29, 35, 74,  8, 44, 11, 69, 44])
testpoints = [mesh[points], f(mesh[points], 1)]
```

```python
model = NP(3, 1, 1)
optimizer = torch.optim.Adam(model.parameters(), lr=3e-3)
```

```python
def sample_plot(np, context, linspace):
    context_x = context[0]
    context_y = context[1]
    
    mu, sigma, log_p, en_dist, t_en_dist, MSE  = np(context_x, context_y, linspace)
    low = mu-sigma
    high = mu+sigma
    plt.plot(linspace.transpose(0,1).squeeze().numpy(), mu.transpose(0,1).squeeze().detach().numpy())
    plt.fill_between(linspace.transpose(0,1).squeeze().numpy(), low.transpose(0,1).squeeze().detach().numpy(), high.transpose(0,1).squeeze().detach().numpy(), facecolor='#65c9f7', interpolate=True)
    plt.scatter(context[0], context[1], c='black')
    
def variance_plot(np, context, linspace):
    context_x = context[0]
    context_y = context[1]
    mu, sigma, log_p, en_dist, t_en_dist, MSE  = np(context_x, context_y, linspace)
    plt.plot(linspace.transpose(0,1).squeeze().numpy(), sigma.transpose(0,1).squeeze().detach().numpy(),  c = 'r')
```

```python
for a in range(1,5):
    model = NP(3, 1, 1)
    optimizer = torch.optim.Adam(model.parameters(), lr=3e-3)
    alpha = 2**a/5
    for i in range(20):
        print(5*i)
        train(datapoints, model, 5, 99, alpha, optimizer)
        sample_plot(model, testpoints, mesh)
        variance_plot(model, testpoints, mesh)
        plt.savefig('./Figs/KLD/alpha{}epoch{}.png'.format(alpha, 5*(i+1)))
        plt.show()
```
