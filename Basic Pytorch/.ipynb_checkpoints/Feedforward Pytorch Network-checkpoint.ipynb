```python
import torch
from matplotlib import pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable

#gnerates 3D data. The network will predict z from x and y.
datapoints = 3000
testpoints = 1000

#the function which will generate the z coordinate from x and y in the data (no outliers for now)
def f(x, y):
    return 2*y**2 + np.cos(5*x)

x_train = []
z_train = []
for i in range(datapoints):
    x_train.append(np.random.uniform(0, 1, size = 2))
    
for i in range(len(x_train)):
    z_train.append(f(x_train[i][0], x_train[i][1]))

x_test = []
z_test = []
for i in range(testpoints):
    x_test.append(np.random.uniform(0, 1, size = 2))
    
for i in range(len(x_test)):
    z_test.append(f(x_test[i][0], x_test[i][1]))    

x_train = np.array(x_train)
z_train = np.array(z_train)
x_test = np.array(x_test)
z_test = np.array(z_test)
```

```python
#a plot of the data

xs = x_train[:,0]
ys = x_train[:,1]
zs = z_train

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c='r', marker = '.')
```

```python
class Feedforward(nn.Module):
    
    def __init__(self, input_size, hidden_size_1, hidden_size_2, hidden_size_3):
        super(Feedforward, self).__init__()
        self.input_size = input_size
        self.hidden_size_1 = hidden_size_1
        self.hidden_size_2 = hidden_size_2
        self.hidden_size_3 = hidden_size_3
        self.fc1 = torch.nn.Linear(self.input_size, self.hidden_size_1)
        self.relu1 = torch.nn.ReLU()
        self.fc2 = torch.nn.Linear(self.hidden_size_1, self.hidden_size_2)
        self.relu2 = torch.nn.ReLU()
        self.fc3 = torch.nn.Linear(self.hidden_size_2, self.hidden_size_3)
        self.relu3 = torch.nn.ReLU()
        self.fc4 = torch.nn.Linear(self.hidden_size_3, 1)
        #self.sigmoid = torch.nn.Sigmoid()
        
    def forward(self, x):
        hidden_1 = self.fc1(x)
        relu_1 = self.relu1(hidden_1)
        hidden_2 = self.fc2(relu_1)
        relu_2 = self.relu2(hidden_2)
        hidden_3 = self.fc3(relu_2)
        relu_3 = self.relu3(hidden_3)
        output_1 = self.fc4(relu_3)
        #output = self.sigmoid(output_1)
        return output_1
        
```

```python
network = Feedforward(2, 10, 10, 10)
criterion = torch.nn.MSELoss()
optimizer = torch.optim.SGD(network.parameters(), lr=0.1)

x_train = torch.from_numpy(x_train)
z_train = torch.from_numpy(z_train)
x_test = torch.from_numpy(x_test)
z_test = torch.from_numpy(z_test)

print(x_train)
```

```python
network.eval()
z_pred = network.forward(x_test.float())
before_train = criterion(z_pred.squeeze(), z_test.float())
print('Test loss before training', before_train.item())
```

```python
z_test_pred = network.forward(x_test.float())

initial_x_test = x_test.detach().numpy()
initial_z_test_pred = z_test_pred.detach().numpy()

xs = initial_x_test[:,0]
ys = initial_x_test[:,1]
zs = initial_z_test_pred

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c='r', marker = '.')

print(network)
```

```python
network.train()
epoch = 5
for epoch in range(epoch):
    for i in range(len(x_train)):
        X, Y = x_train[i], z_train[i]
        X, Y = Variable(torch.FloatTensor(X.float()).float(), requires_grad=True), Variable(torch.FloatTensor([Y.float()]).float(), requires_grad=False)
        optimizer.zero_grad()
        outputs = network(X)
        loss = criterion(outputs, Y)
        loss.backward()
        optimizer.step()
        if i % 300 == 0:
            print("Epoch {} - loss: {}".format(epoch, loss.data))
```

```python
network.eval()
z_pred = network(x_test.float())
after_train = criterion(z_pred.squeeze(), z_test.float())
print('Test loss after Training', after_train.item())
```

```python
z_test_pred = network.forward(x_test.float())

x_test = x_test.detach().numpy()
z_test_pred = z_test_pred.detach().numpy()

xs = x_test[:,0]
ys = x_test[:,1]
zs = z_test_pred

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c='r', marker = '.')
```
