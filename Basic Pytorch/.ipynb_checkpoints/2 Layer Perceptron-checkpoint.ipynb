```python
import torch
import numpy
from matplotlib import pyplot as plt
import torch.nn.functional as F

num_points = 50
X = np.random.uniform(low=-50, high=50, size=(num_points, 1))
Y = X + np.random.normal(size=(num_points, 1), scale = 5)

plt.scatter(X, Y)
```

```python
input_dimension, hidden_size, output_dimension = 1, 10, 1

x = torch.from_numpy(X).float()
y = torch.from_numpy(Y).float()
w1 = torch.randn(input_dimension, hidden_size, requires_grad=True)
w2 = torch.randn(hidden_size, output_dimension, requires_grad=True)
learning_rate = 1e-6

for t in range(100):
    y_pred = F.relu(x.mm(w1)).mm(w2)
    loss = (y_pred - y).pow(2).sum()
    print(t, loss.item())
    loss.backward()
    
    with torch.no_grad():
        w1 -= learning_rate * w1.grad
        w2 -= learning_rate * w2.grad
        
        w1.grad.zero_()
        w2.grad.zero_()
```

```python
Xtest = np.linspace(-50, 50, 50)[np.newaxis].transpose()
Xin = torch.from_numpy(Xtest).float()
Ycalc = F.relu(Xin.mm(w1)).mm(w2)

Xplot = Xin.detach().numpy()
Yplot = Ycalc.detach().numpy()
plt.scatter(Xplot, Yplot)
plt.scatter(X, Y)
```
